## **2 - Training and testing splitting plus correlations**

```python
# training - testing split: 4 years to callibrate the correlations, 1 year to test performance

yf_returns_train = yf_returns.iloc[0:2150,:]
print(yf_returns_train.shape)
yf_returns_test = yf_returns.iloc[2150:2515,:]
print(yf_returns_test.shape)
```

The data has missing values. In general we would need to treat them, either deleting them or replacing them with the previous value, the  median, the mean, etc. Since we are goign to use the corr() function we don't need to do that: Any NaN values are automatically excluded

```python
# Generate the correlation matrix of this data
corr = yf_returns_train.corr()
corr
```

```python
# Generate data for optimization with AMPL ---> copy this data into the optimization notebook

f = open("IndexConstruction.dat", "w")
f.write('param q := 7;\n\n')
f.write('set tickers := ')
for i in tickerList:
    f.write('%s ' % i)
f.write('; \n\n')
f.write('param corr: ')
for i in tickerList:
    f.write('%s ' % i)
f.write(':= \n')
for i in tickerList:
    f.write('%s ' % i)
    for j in tickerList:
        f.write('%.3f ' % corr[i][j])
    f.write('\n')
f.write(';')
f.close()
```

## **4 - Obtaining the market cap to complete the index construction, with the outputs of the AMPL optimization**

```python
# get the market capitalization of the companies and transform it into a dictionary
Mcap = pd.read_excel('MarketCaps.xlsx')
market_cap = pd.Series(Mcap.Value.values,index=Mcap.Tickers).to_dict()
```

```python
# Copy and paste the output of AMPL here
assign = {
'ACN': ['AAPL','ABBV','ABT','ACN','ADBE','AMD','AVGO','AXP','BAC','BLK','BMY','CL','COST','DE','DHR','DUK','EXC','GOOG','GOOGL','GS','HD','JPM','KO','LIN','LLY','LOW','MDLZ','MS','MSFT','NEE','NKE','NVDA','ORCL','PEP','PFE','PG','QCOM','SCHW','SO','TGT','TMO','TSLA','TXN','UNH','UNP','UPS',],
'DOW': ['CAT','COP','CVS','DOW','EMR','F','FDX','GE','IBM','KHC','MET','MO',],
'GILD': ['GILD',],
'USB': ['BA','BK','BKNG','C','COF','CVX','GD','GM','MMM','RTX','USB',],
'V': ['AMGN','AMT','AMZN','CHTR','CMCSA','CRM','CSCO','DIS','HON','INTC','JNJ','LMT','MA','MCD','MDT','META','MRK','NFLX','PYPL','SBUX','T','TMUS','V','VZ','WMT',],
'WFC': ['AIG','SPG','WFC','XOM',],
}
```

```python
# Determine the normalized market capitalization of tickers assigned to each selected stock

tot_assign = 0
market_cap_assign = {}
for j in assign:
    tot = 0
    for i in assign[j]:
        tot += market_cap[i]
    market_cap_assign[j] = tot
    tot_assign += tot

for j in market_cap_assign:
    market_cap_assign[j] = market_cap_assign[j]/tot_assign
    print('%s\t%.4f' % (j,market_cap_assign[j]))
```

```python
# The returns of each of the components of the index during the quarters of interest
return_qtr_1 = {i:(yf_returns_test[i][92] - yf_returns_test[i][1])/yf_returns_test[i][1] for i in assign}
return_qtr_2 = {i:(yf_returns_test[i][183] - yf_returns_test[i][1])/yf_returns_test[i][1] for i in assign}
return_qtr_3 = {i:(yf_returns_test[i][274] - yf_returns_test[i][1])/yf_returns_test[i][1] for i in assign}
return_qtr_4 = {i:(yf_returns_test[i][364] - yf_returns_test[i][1])/yf_returns_test[i][1] for i in assign}
```

```python
# The returns of the index after after 1 QTR, 2 QTRs, 3 QTRs, 4 QTRs
index_weighted_return_qtr_1 = 0
index_weighted_return_qtr_2 = 0
index_weighted_return_qtr_3 = 0
index_weighted_return_qtr_4 = 0
for group_name in assign:
    index_weighted_return_qtr_1 += market_cap_assign[group_name] * return_qtr_1[group_name]
    index_weighted_return_qtr_2 += market_cap_assign[group_name] * return_qtr_2[group_name]
    index_weighted_return_qtr_3 += market_cap_assign[group_name] * return_qtr_3[group_name]
    index_weighted_return_qtr_4 += market_cap_assign[group_name] * return_qtr_4[group_name]

# Print the weighted returns per quarter
print("Weighted return after 1 QTR:", index_weighted_return_qtr_1)
print("Weighted return after 2 QTRs:", index_weighted_return_qtr_2)
print("Weighted return after 3 QTRs:", index_weighted_return_qtr_3)
print("Weighted return after 4 QTRs:", index_weighted_return_qtr_4)
```

```python
# Print the ratios between the returns of the index and the benchmark after 1QTR, 2QTRs, 3 QTRs and 4QTRs. Numbers should be as close to 1 as possible!
print("Ratio after 1 QTR:", index_weighted_return_qtr_1 / benchmark[0])
print("Ratio after 2 QTR:", index_weighted_return_qtr_2 / benchmark[1])
print("Ratio after 3 QTR:", index_weighted_return_qtr_3 / benchmark[2])
print("Ratio after 4 QTR:", index_weighted_return_qtr_4 / benchmark[3])
```
